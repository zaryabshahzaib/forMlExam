# Bagging 

```python
# let's make the basic imports

import numpy as np
import pandas as pd

import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['figure.dpi']= 180 # set the resolution to x dpi
```

## Iris dataset for basic testing of bagging on classification tasks

```python
from sklearn import datasets 
iris = datasets.load_iris()
```

```python
#print(iris['DESCR'])
```

```python
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['Iris species'] = iris.target
iris_df
```

```python
#import seaborn as sb

#sb.set(style="ticks", color_codes=True)
#g = sb.pairplot(iris_df, hue="Iris species")
```

## Split features and targets, store names

```python
X = iris_df[iris.feature_names].values
Y = iris_df['Iris species'].values

# let's store the feature names in compact lists, for plotting
feature_names = iris.feature_names
feature_names = [x.replace(' (cm)', '') for x in feature_names]

#https://gist.github.com/curran/a08a1080b88344b0c8a7

class_names_dict = {0:'Setosa', 1:'Versicolor', 2:'Virginica'}
class_names = ['Setosa', 'Versicolor', 'Virginica']
```

## Use only two features (to make it more challenging and make visualization being fully informative)

Let's consider only the first two features (try out with different selections of two features!)

```python
from sklearn.model_selection import train_test_split

f1_name = iris.feature_names[0]
f2_name = iris.feature_names[1]

X_train, X_test, Y_train, Y_test = train_test_split(X[:, 0:2], Y, 
                                                    test_size=0.25, random_state=99)
```

```python
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

## Set and train a bagging classifier

```python
from sklearn.ensemble import BaggingClassifier
```

Method `BaggingClassifier()` creates a bagging object for classification. 

A number of keyword arguments are available to guide or constrain bagging process.

`base_estimator=None, 
n_estimators=10, 
max_samples=1.0, 
max_features=1.0, 
bootstrap=True, 
bootstrap_features=False, 
oob_score=False, 
warm_start=False, 
n_jobs=None, 
random_state=None, 
verbose=0`

- `base_estimator` : object or None, optional (default=None)<br>
    The base estimator to fit on random subsets of the dataset.<br>
    If None, then the base estimator is a decision tree.<br><p>
 
- `n_estimators` : int, optional (default=10)<br>
    The number of base estimators in the ensemble.<br><p>

- `max_samples` : int or float, optional (default=1.0)<br>
    The number of samples to draw from X to train each base estimator.<br><p>
    
    - If int, then draw `max_samples` samples.
    - If float, then draw `max_samples * X.shape[0]` samples.<br><p>
 
- `max_features` : int or float, optional (default=1.0)<br>
    The number of features to draw from X to train each base estimator.<br><p>
    - If int, then draw `max_features` features.<br>
    - If float, then draw `max_features * X.shape[1]` features.


```python
#help(BaggingClassifier)
```

## Select the base model

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier

base_model = 'dt'

if base_model == 'dt':
    criterion = 'entropy' # 'gini
    max_depth = None
    classifier = DecisionTreeClassifier(criterion=criterion, max_depth=max_depth,
                                        random_state=999)
elif base_model == 'knn':
    n_neighbors = 5
    weights = 'uniform'
    classifier = KNeighborsClassifier()
else:
    print('Select a known base model!')

max_samples  = 0.4  # must be float to be a fraction!
max_features = 1.0
n_estimators = 9

bagging = BaggingClassifier(classifier, n_estimators=n_estimators,
                            max_samples=max_samples, max_features=max_features,
                            random_state=1234)
```

## Train and test the bagging model

```python
bagging.fit(X_train, Y_train)
```

```python
from sklearn import metrics 

Y_pred = bagging.predict(X_test)

accuracy = metrics.accuracy_score(Y_test, Y_pred)

print('Accuracy of bagging: {:.2f}'.format(accuracy))
```

## Confusion matrix

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

def show_confusion_matrix(true_labels, learned_labels, class_names):

    cmat = confusion_matrix(true_labels, learned_labels) 

    plt.figure(figsize=(12, 2.5))

    plt.tick_params(labelsize=8)

    
    hm = sns.heatmap(cmat.T, square=True, annot=True, fmt='d', cbar=True,
                     xticklabels=class_names,
                     yticklabels=class_names, 
                     cmap="seismic", 
                     annot_kws={"size":12}, cbar_kws={'label': 'Counts'})

    # this is to set the last axis of the figure, the colorbar in this case
    hm.figure.axes[-1].yaxis.label.set_size(10) # fontsize of bar label
    hm.figure.axes[-1].tick_params(labelsize=8) # fontsize of ticks labels
    
    #plt.ylim(10, 0)

    plt.xlabel('True label', fontsize=9)
    plt.ylabel('Predicted label', fontsize=9)
    
    plt.show()
    
show_confusion_matrix(Y_test, Y_pred, class_names)
```

## Performance of the base model alone

```python
classifier_nobagging = classifier

classifier_nobagging.fit(X_train, Y_train)

Y_pred = classifier_nobagging.predict(X_test)

accuracy_nobagging = metrics.accuracy_score(Y_test, Y_pred)

print('Accuracy of base estimator: {:.2f}'.format(accuracy_nobagging))
```

## List and visualize trained bagging estimators vs. bagging estimator

```python
bagging.estimators_
```

We have considered only two features, such that we can effectively visualize the decision boundaries in the feature space.

At this aim we can reuse the plot_decision_regions() function of the last notebook. It expects a two-dimensional feature set as a numpy array, `X_two_f`.

Note that if the the original feature array `X` has n features, any two of them, let's say `a` and `b`, can be selected using the form `X[:, [a,b]]`.

```python
#Create a custom color map
from matplotlib.colors import ListedColormap
cmap_clear = ListedColormap(['orange', 'lightgray', 'lime'])
    
def plot_decision_regions(X_two_f, Y, estimator, subplot = None, title='', 
                          x1_name='', x2_name='', 
                          cmap='viridis', size=(14,8), data=True):
    
    # get feature ranges for creating a meshgrid
    feature_ranges = np.array([ (np.floor(np.min(X_two_f[:, i])), 
                                 np.ceil(np.max(X_two_f[:, i]))) 
                                 for i in range(2)]  )

    num_of_pts = 600
    
    # a meshgrid of values for (f1, f2), these are the points where 
    # algorithm will be queried and that we want to visualize
    xx, yy = np.meshgrid( np.linspace(feature_ranges[0, 0],
                                      feature_ranges[0, 1], num=num_of_pts),
                          np.linspace(feature_ranges[1, 0],
                                      feature_ranges[1, 1], num=num_of_pts))
    #print(xx.shape, yy.shape)

    # get the labels of all the points in the meshgrid
    labels_mesh = estimator.predict(np.c_[xx.ravel(), yy.ravel()])

    # reshape the array to be of the same dimension as xx and yy
    labels_mesh = labels_mesh.reshape(xx.shape)

    #print(np.c_[xx.ravel(), yy.ravel()].shape, labels_mesh.shape)

    if subplot == None:
        plt.figure(figsize=size)
        plt.title(title)

        # the color map automatically assigns a different color to the 
        # different values of labels 
        plt.pcolormesh(xx, yy, labels_mesh, cmap=cmap)

        # Add the original data points 
        scatter = plt.scatter(X_two_f[:, 0], X_two_f[:, 1], c=Y, s=50, 
                              edgecolor='white', cmap=cmap)
        plt.colorbar(scatter)

        plt.xlabel(x1_name)
        plt.ylabel(x2_name)
        
        plt.show()
        
    else:
        # the color map automatically assigns a different color to the 
        # different values of labels  
        subplot.pcolormesh(xx, yy, labels_mesh, cmap=cmap)

        # Add the original data points 
        if data:
            scatter = subplot.scatter(X_two_f[:, 0], X_two_f[:, 1], c=Y, s=50, 
                                      edgecolor='white', cmap=cmap)
            #subplot.set_colorbar(scatter)

        subplot.set_xlabel(x1_name)
        subplot.set_ylabel(x2_name)
        
```

```python
plot_decision_regions(X_train, Y_train, bagging, None, 
                      'Bagging DT - Accuracy: {:.2f}'.format(accuracy), 
                      f1_name, f2_name, cmap=cmap_clear)
```

Above is the plot of the decision regions of the bagging classifier. Let's check how it compares to the individual trained models on the bagging estimator.

```python
def visualize_estimators(X, Y, estimators, n=None, title='', 
                         x1_name='', x2_name='',
                         cmap = 'viridis', data=True):
                         
    if n == None:
        n = len(estimators)
        
    fig = plt.figure(figsize=(15, 12)) 
    fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.2, wspace=0.15) 
    
    ncols = 3
    nrows = int(np.ceil(n / ncols))

    for i in range(n): 
        subplot = fig.add_subplot(nrows, ncols, i+1) 
        plot_decision_regions(X_train, Y_train, estimators[i], subplot, '', 
                              f1_name, f2_name,
                              cmap=cmap, data=data)

   
    plt.show()
```

```python
visualize_estimators(X_train, Y_train, bagging.estimators_, title='Bagging DTs', 
                     x1_name = f1_name, x2_name = f2_name, 
                     cmap = cmap_clear, data = True)
```

## Visualize decision boundaries

For visualizing the decision boundaries we can only use two features. This means that the we need to train on two features only. 

An effective way of doing this is to **loop over all possible combinations of two features and display the relative decision boundaries.**

At this aim we can reuse the `plot_decision_regions()` function that we have used above.

```python
#Create a custom color map
from matplotlib.colors import ListedColormap
cmap_clear = ListedColormap(['orange', 'lightgray', 'lime'])
    
def plot_decision_regions(X_two_f, Y, estimator, subplot = None, title='', 
                          x1_name='', x2_name='', 
                          cmap='viridis', size=(14,8), data=True):
    
    # get feature ranges for creating a meshgrid
    feature_ranges = np.array([ (np.floor(np.min(X_two_f[:, i])), 
                                 np.ceil(np.max(X_two_f[:, i]))) 
                                 for i in range(2)]  )

    num_of_pts = 600
    
    # a meshgrid of values for (f1, f2), these are the points where 
    # algorithm will be queried and that we want to visualize
    xx, yy = np.meshgrid( np.linspace(feature_ranges[0, 0],
                                      feature_ranges[0, 1], num=num_of_pts),
                          np.linspace(feature_ranges[1, 0],
                                      feature_ranges[1, 1], num=num_of_pts))
    #print(xx.shape, yy.shape)

    # get the labels of all the points in the meshgrid
    labels_mesh = estimator.predict(np.c_[xx.ravel(), yy.ravel()])

    # reshape the array to be of the same dimension as xx and yy
    labels_mesh = labels_mesh.reshape(xx.shape)

    #print(np.c_[xx.ravel(), yy.ravel()].shape, labels_mesh.shape)

    if subplot == None:
        plt.figure(figsize=size)
        plt.title(title)

        # the color map automatically assigns a different color to the 
        # different values of labels 
        plt.pcolormesh(xx, yy, labels_mesh, cmap=cmap)

        # Add the original data points 
        scatter = plt.scatter(X_two_f[:, 0], X_two_f[:, 1], c=Y, s=50, 
                              edgecolor='white', cmap=cmap)
        plt.colorbar(scatter)

        plt.xlabel(x1_name)
        plt.ylabel(x2_name)
        
        plt.show()
        
    else:
        # the color map automatically assigns a different color to the 
        # different values of labels  
        subplot.pcolormesh(xx, yy, labels_mesh, cmap=cmap)

        # Add the original data points 
        if data:
            scatter = subplot.scatter(X_two_f[:, 0], X_two_f[:, 1], c=Y, s=50, 
                                      edgecolor='white', cmap=cmap)
            #subplot.set_colorbar(scatter)

        subplot.set_xlabel(x1_name)
        subplot.set_ylabel(x2_name)
        
```

The function `show_regions_using_all_feature_pairs()` creates a matrix of decision regions, considering all pairwise combinations of features. A new model is trained using the selected pair and the decision regions are plotted accordingly. Also the accuracy of the model is reported in the plots.

The function takes as input *any* estimator, specified by the `estimator` parameter, and allows to set the values of the parameters of the estimator through the dictionary argument `estimator_params`. Note the use of the syntax `**estimator_params` in the call of `estimator()`, that expands the dictionary, mapping the keys to keyword arguments of the estimator function object, allowing in this way to pass named arguments to a generic function.

```python
def show_decision_regions_for_all_feature_pairs(X_train, Y_train, X_test, Y_test,
                                                estimator = None,
                                                estimator_params = {'n_estimators':100, 'random_state':None}, 
                                                feature_names = None, #class_names = None, 
                                                ncols = 3, 
                                                size = (15,9), 
                                                cmap=cmap_clear):  
    import itertools

    fig = plt.figure(figsize=size) 
    fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.25, wspace=0.18) 
    
    n_plots = len(list(itertools.combinations(list(range(X_train.shape[1])), 2)))

    nrows = int(np.ceil(n_plots / ncols))
    
    #n_estimators = parameters['n_estimators'] 
    
    for i,ff in enumerate(itertools.combinations(list(range(X_train.shape[1])), 2)):

        classifier = estimator(**estimator_params) #random_state=rnd_seed, 
        
        clf_name = classifier.__class__.__name__

        X_two_features_train = X_train[:, [ff[0], ff[1]] ].copy()
        X_two_features_test = X_test[:, [ff[0], ff[1]] ].copy()
    
        classifier.fit(X_two_features_train, Y_train)

        Y_pred = classifier.predict(X_two_features_test)
    
        accuracy = metrics.accuracy_score(Y_test, Y_pred)

        print('Features:', ff)
        #print(X_two_features_train.shape, X_two_features_test.shape)
    
        subplot = fig.add_subplot(nrows, ncols, i+1) 
    
        plot_decision_regions(X_two_features_train, Y_train, classifier, subplot, 
                              '', feature_names[ff[0]], feature_names[ff[1]],
                              cmap = cmap)
    
        title = '{} [{}, {}] - Accuracy: {:.2f}'.format(clf_name, 
                                                        feature_names[ff[0]], feature_names[ff[1]], 
                                                        accuracy)
        
        subplot.set_title(title, color='black', fontsize=8)
    
    plt.show()
```

```python
show_decision_regions_for_all_feature_pairs(X_train, Y_train, X_test, Y_test,
                                            estimator=AdaBoostClassifier,
                                            estimator_params = {'n_estimators':50, 'random_state':99}, 
                                            feature_names = feature_names)
```

## Search for best cross-validated number of estimators

What is the best number of estimators? What is the best *weak* base model? 

We can try to answer these questions by performing some search in the parameter space.

```python
from sklearn import model_selection as ms

# the max_depth parameter of the base_estimator (__) has to be set, otherwise
# GridSearchCV makes use of the default value for DT (None). 
# the base_estimator has to be given in the constructor of ABC
#
parameters_grid = {'base_estimator__max_depth': [1],
                    'n_estimators': [1, 2, 10, 25, 50, 75, 100, 125, 150] }

estimator = ms.GridSearchCV(AdaBoostClassifier(base_estimator=DecisionTreeClassifier()), 
                                         param_grid=parameters_grid)

estimator.fit(X_train, Y_train)
```

```python
print('Best CV accuracy score: {:.3f}'.format(estimator.best_score_))

print('Best estimator: {}'.format(estimator.best_estimator_))

print('\nBest number of base models: {}'.format(estimator.best_params_['n_estimators']))
```

```python
#Predict the response for test dataset using the best estimator
Y_pred = estimator.predict(X_test)

accuracy = metrics.accuracy_score(Y_test, Y_pred)

print('Accuracy on testing: {:.2f}'.format(accuracy))
```
