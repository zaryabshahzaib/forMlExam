# Regression problems



## Linear Regression




***
<font size="+1">

- What is the meaning of **<i>linear</i>**?
  
  The prediction value $\hat{y}$ is obtained by **summing up the contribution of each input feature** $x_i, \ i=1, \ldots, d$ in ${\bf x}$<br><p>
    
  - E.g., Predict the price $y({\bf x})$ of an apartment based on four predictor features 
    
    $${\bf x} = (square\ meters,\ floor,\ number\ of\ rooms,\ house\ tax)?$$<br> 
    $\rightarrow$ In a linear model the contribution of each feature *sums up*<br><p>
    If ${\bf x} = (130.5, 12, 6, 2210.63)$, the linear predictor says that:<br><p>
    $$y({\bf x}) = 130.5w_1 + 12w_2 + 6w_3 +  2210.63w_4$$
        
</font>

***

***
<font size="+1">
    
- What is the meaning of ${\bf w}$? <br><p>

  ${\bf w}$ is a **numeric vector weighting the contribution of each input feature** $x_i, \ i=1, \ldots, d$, in ${\bf x}$, in the summation making up the prediction $\hat{y}$ <br><p>
           
</font> 

***

<font size="+1">
    
- If ${\bf x} = (130.5, 12, 6, 2210.63)$, <br><p>
- and the learned weights are ${\bf w} = (20245.7,\ 374.6,\ 78.5,\ 104.4, 55.1)$, the linear predictor says that:<br><p>
    
$$\hat{y}({\bf x}) = 20245.7 + 130.5\times 374.6 + 12\times 78.5 + 6\times 104.4 + 2210.63\times 55.1 = 192505.11$$
    
</font> 

```python
import numpy as np

#w = np.array([1.5, 0.4, 0.9, 1.1])
w = np.array([20245.7, 374.6, 78.5, 104.4, 55.1])
x = np.array([1, 130.5, 12, 6, 2210.63])

price = x.dot(w)
print('Predicted house price: {:.2f}'.format(price))
```

## What if the envisaged relationship isn't linear?

<font size="+1">
    
What if the model to represent the relation between features and predictions is not linear?
    
Check if we can apply **transformations to the variables to make it linear.** 
</font>

<font size="+1">
For instance, if the envisaged model is of the following form:<br>

$$y(t) = ae^{bt}$$

e.g., $t$ is the day of the start of the infection of corona virus and $y(t)$ is the number of infected people at day $t$.

this is clearly not a linear, rather **exponential,** relation between predictor variables and predictions!
</font>

<font size="+1">
Applying a $\log()$ transformation:
    
$$\log(y) = \log(a) + bt$$
</font>

<font size="+1">
This is still not linear for the relation between $t$ and $y$, since $y$ appears as a logarithm, but we can make it linear by applying the transformation $y' = \log(y)$:
    
   $$y' = \log(a) + bt$$
    
Now it's linear between predictor and prediction variables, but it is still <u>not linear with respect to the parameters that we have to estimate,</u> $a$ and $b$, since $a$ appears with a logarithm.

However, we can we operate a further transformation, $a' = \log(a)$ such that now the relation becomes fully linear and we can safely apply the methods of linear regression
    
$$y' = a' + bt$$
    
Of course, in order to recover the values of the original exponential relation, we have to apply all the used transformations.
</font>

## Numeric example: log transformation to exponential

<font size="+0.8">
    
Applying a $log()$ transformation to exponential data indeed **linearize the data.**

This can be seen in the numeric example below, where $n$ points are generated from an exponential function *corrupted* by some Gaussian noise (*homoscedastic,* uniform for all points).

The plots show first the original data and underlying reference function, and then transformed data and the new (linear) reference function.
</font>

```python
import numpy as np

n = 50
v_max = 10
v_min = 5

# generate n points uniformly distributed between v_min and v_max
t = v_min + (v_max - v_min) * np.random.random((n))

# some arbitrarily defined exponential function as a reference
a = 3.5
b = 0.5
y_ref = a * np.exp(b * t) 

# some Gaussian noise, homoscedastic, to corrupt the reference values
sigma = 15
mu = 5
y = y_ref + (sigma * np.random.randn(n) + mu)
```

```python
import matplotlib.pyplot as plt

# plot data points
plt.figure(figsize=(7,6))
plt.xlabel('t')
plt.ylabel('y(t)')
plt.scatter(t, y, s=4)

# plot reference function
x = np.linspace(v_min, v_max, 5000)
plt.plot(x, a * np.exp(b * x), color='r')
plt.show()
```

```python
# transform parameters and data according to the (know) map
alog = np.log(a)
blog = b
ylog = np.log(y)

# plot data points
plt.figure(figsize=(7,6))
plt.xlabel('t')
plt.ylabel('$\log(y(t))$')
plt.scatter(t, ylog, s=4)

# plot reference function
x = np.linspace(v_min, v_max, 5000)
plt.plot(x, alog + blog * x, color='r')

plt.show()
```

## Ordinary Least Squares (OLS)

<span style="font-size:140%">Given a dataset with $n$ examples<br> $$\mathcal{D} = \{y^{(i)}, {\bf x}^{(i)}\}_{i=1}^n$$<br>
    where the explanatory variables ${\bf x}^{(i)}$ are <u>multivariate,</u> ($d$-dimensional) and the predictions $y^{(i)}$ are <u>scalars,</u> the **multiple linear regression problem** consists in: 
</span>


***
<font size="+1">
    
**Ordinary Least Squares (OLS):** We solve the linear regression problem by <u>minimizing the squares of the residuals / errors</u>
    
 $$\ell({\bf w}) = \frac{1}{m}\sum_{i=1}^n \Big(y^{(i)} - {\bf w}\cdot{\bf x}^{(i)}\Big)^2 =
 \frac{1}{m} \sum_{i=1}^n \Big(y^{(i)} - \hat{y}^{(i)}\Big)^2.$$
   
</font>

***

<div style="text-align: left;">
  
</div>

***
<font size="+1">
Adopting a compact <u>matrix form:</u>
    
$$\ell({\bf w}) = ({\bf Y} - \hat{\bf Y})^T({\bf Y} - \hat{\bf Y})$$    

- ${\bf Y}$: vector of **true labels**<br><p>
    $$
    {\bf {Y}} = 
    \begin{pmatrix}
    {y}^{(1)}\\ {y}^{(2)}\\ \vdots \\ {y}^{(n)}
    \end{pmatrix}
    $$
<br><p>
- $\hat{{\bf Y}}$: vector of **predicted labels** based on the weights, ${\bf w}^*$ 
    $$
    {\bf \hat{Y}} = 
    \begin{pmatrix}
    \hat{y}^{(1)}\\ \hat{y}^{(2)}\\ \vdots \\ \hat{y}^{(n)}
    \end{pmatrix}
    $$
</font>

```python
#https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_regression.html
    
import sklearn.datasets as skd
import matplotlib.pyplot as plt

# generate a linear regression dataset
X, Y = skd.make_regression(n_samples=200, n_features=1, noise=15, random_state=1235)

# plot regression dataset
plt.figure(figsize=(7,6))
plt.xlabel('x')
plt.ylabel('y')
plt.scatter(X, Y, s=2)
plt.show()
```

```python
from sklearn import linear_model 

# Create a lineear regression object
OLS = linear_model.LinearRegression(fit_intercept=False) 

# Train the model using the training set, features and true labels
OLS.fit(X, Y)

# Get the predicted labels on the input features of the training set
Yhat = OLS.predict(X)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# The mean rooted squared error
print('Root Mean Square Error: {:.2f}'.format(np.sqrt(mean_squared_error(Y, Yhat))))

# The coefficient of determination: 1 is perfect prediction
print('Coefficient of determination (R2): {:.2f}'.format(r2_score(Y, Yhat)))
      
```

```python
# plot regression dataset
plt.figure(figsize=(7,6))
plt.xlabel('x')
plt.ylabel('y')
plt.scatter(X, Y, s=2)

# plot learned regression line
x_plot = np.linspace(X.min(), X.max(), 1000)

OLS_y = OLS.intercept_ + OLS.coef_ * x_plot
plt.scatter(x_plot, OLS_y, color='r', s=0.3)

plt.show()
```

***
<font size="+1">
    Summarizing, looking for a <u>linear regression model based on OLS,</u> is equivalent to say that:<br><p>
    
- *Given* the labeled dataset $\mathcal{D} = \{y^{(i)}, {\bf x}^{(i)}\}_{i=1}^n$<br><p><p>
    
- *Assumed* (inductive bias) that for an input vector ${\bf x} = (1, x_1, x_2, \ldots x_d)^T$, the associated value $y$ can be reliably predicted by applying a <u>linear model</u> of the type $\hat{y} = {\bf w}^T\cdot {\bf x}$<br><p>
    
    - Where ${\bf w}$ is a numeric vector weighting the contribution of each input feature $x_i, \ i=1, \ldots, d$, in ${\bf x}$.<br><p><p>
    
- $\rightarrow$ *Find* the values ${\bf w}^*$ of the weights  such that the SSE, $\ell({\bf w})$, over the dataset $\mathcal{D}$ is minimized:<br><p>
    $${\bf w}^* = \arg\min_{\bf w} \ell({\bf w}) = \arg\min_{\bf w}  \sum_{i=1}^n \Big(y^{(i)} - {\bf w}\cdot{\bf x}^{(i)}\Big)^2 = \arg\min_{\bf w}  \sum_{i=1}^n \Big(y^{(i)} - \hat{y}^{(i)}\Big)^2.$$
</font>
***

<font size="+1">
<span style="color:red">The choice ${\bf w}^*$ is such that $\ell({\bf w}^*)$ is the minimal value of $\ell({\bf w})$ over the training dataset $\mathcal{D}$. </span>
    
This is equivalent to say that the predictions $\hat{{\bf Y}}$ in $\mathcal{D}$ made using ${\bf w}^*$, 
    
$$ 
{\bf \hat{Y}} = 
\begin{pmatrix}
\hat{y}^{(1)}\\ \hat{y}^{(2)}\\ \vdots \\ \hat{y}^{(n)}
\end{pmatrix} = 
%
\begin{pmatrix}
w_1^* \\ w^*_2 \\ \vdots \\ w^*_n
\end{pmatrix}^T
%
\cdot
%
\begin{pmatrix}
x_1^{(1)}& x_2^{(1)}& \ldots & x_d^{(1)}\\
x_1^{(2)}& x_2^{(2)}& \ldots & x_d^{(2)}\\
\ldots & \ldots & \ldots & \ldots \\
x_1^{(n)}& x_2^{(n)}& \ldots & x_d^{(n)}
\end{pmatrix}
%
= {\bf w}^{*^{^T}}\cdot {\bf X}
$$
    
provide the minimal SSE loss in reference to the prediction labels ${\bf Y}$: <br><p>
    $$\ell({\bf w}^*) = ({\bf Y} - \hat{\bf Y})^T({\bf Y} - \hat{\bf Y})$$
    
is minimal over all possible choices of the weights ${\bf w}$.
</font>

## Solution of OLS problems<br>
<font size="+1">
In the Lecture of Feb 19,<br>
    
https://web2.qatar.cmu.edu/~gdicaro/15488/lectures/488-S20-14-FeatureEngineering_Numeric_2.ipynb

we have briefly discussed how a linear regression problem is solved (without digging into the math). Please refer to that lecture for additional details.
    
Here we should remember that the minimization problem 
     $${\bf w}^* = \arg\min_{\bf w} \ell({\bf w}) = \arg\min_{\bf w}  \sum_{i=1}^n \Big(y^{(i)} - {\bf w}\cdot{\bf x}^{(i)}\Big)^2 .$$
is solved by: 

 1. computing the **gradient vector** (i.e., the partial derivatives) of the the loss function $\ell({\bf w})$ with respect to the variables $w_i, i=0, \ldots, d$, and equating each partial derivative to zero:
    
    $$
    \begin{cases}
    \begin{aligned}
    \frac{\partial \ell({\bf w})}{\partial w_0} = 0 \\[1ex]
    \frac{\partial \ell({\bf w})}{\partial w_1} = 0 \\
    \ldots\\
    \frac{\partial \ell({\bf w})}{\partial w_d} = 0 
    \end{aligned}
    \end{cases}
    $$
 
 
 2. finding the values ${\bf w}^*$ of the components of the ${\bf w}$ vector such that all $d$ partial derivatives equations are satisfied.<br><p>


</font>

***
<font size="+1">

<span style="color:red">Once we proceed with the calculations, $\ldots$, we would find that the solution of the OLS optimization problem can be obtained in <u>closed form</u> as:$\ $<br><p>
    
$${\bf w}^* = ({\bf X}^T {\bf X})^{-1}{\bf X}^T{\bf Y}$$
</span>

</font>

***

<font size="+1">
$\Rightarrow$ We can solve an OLS regression problem by matrix transposition, inversion, and multiplication!
</font>

<font size="+1">
    <u>Dimensions (including bias in the ${\bf X}$ matrix):</u><br>
    
- ${\bf X}$: $n\times (d+1)$ matrix<br><p>
- ${\bf X}^T$: $(d+1)\times n$ matrix<br><p>
- ${\bf Y}$: $n$ vector (or, $n\times 1$ matrix)<br><p>
- ${\bf w}$: $(d+1)$ vector (or, $(d+1)\times 1$ matrix)<br><p><p>
    
The equation ${\bf w}^* = ({\bf X}^T {\bf X})^{-1}{\bf X}^T{\bf Y}$ in dimensional form:<br><p>
    $$[d+1] = [(d+1)\times (d+1)]\cdot [(d+1)\times n] \cdot [n]$$
</font>

***
<font size="+1">

<span style="color:red"><u>Watch out!</u></span>
        
The above solution results from the equation:<br>
  
$$ ({\bf X}^T {\bf X}){\bf w} = {\bf X}^T{\bf Y} $$

    
$$
\begin{aligned}
&a_{00}w_0 + a_{01}w_1 + a_{02}w_2 + \ldots + a_{0d}w_d = c_0y^{(0)}\\
&a_{10}w_0 + a_{11}w_1 + a_{12}w_2 + \ldots + a_{1d}w_d = c_1y^{(1)}\\
&\cdots \quad \cdots\\
&a_{d0}w_d + a_{d1}w_1 + a_{d2}w_2 + \ldots + a_{dd}w_d = c_dy^{(d)}
\end{aligned}
$$
    
This <u>linear system of $d+1$ equations in of $d+1$ unknowns</u> has a **unique solution**, ${\bf w}^*$, if and only if the matrix $({\bf X}^T {\bf X})$ (of the coefficients $a_{ij}$ above, is **invertible**.
    
If the matrix is *not* invertible, meaning that  it is **singular**, then the system / solution equation has infinite solutions. 
    
<u>Singular matrix</u>: the matrix has not *full rank,* columns are *not linearly independent,* the determinant is zero.
***
</font>

## Numeric example: solution by hand and solution by `sklearn`

Let's look at an example, to grasp a few additional concepts and better understand the role of the feature matrix ${\bf X}$.</font>

Let's use the following small dataset of $n = 10$ example points of 2-variate features that are generated as Gaussian noise about a 2D hyperplane.

```python
import numpy as np
import pandas as pd

import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['figure.dpi']= 180 # set the resolution to x dpi
```

```python
def hyperplane_2D(x, c2=2, c1=3.5, c0=5):
    return c2*x[1] + c1*x[0] + c0

def hyperplane_nD(x, d):
    c = np.random.random(size = d+1)
    print(c.shape)
    return c[1:].dot(x) + c[0]
  
def generate_multivariate_regression_data(f, n=800, dimensions=2, 
                                          xmin=-10, xmax=10, 
                                          seed=123,
                                          visualize=True):
    
    np.random.seed(seed)
    
    # -------- Generate noisy data points ---------
    # random uniform on the x dimensions
    x = xmin + (xmax - xmin)* np.random.random((dimensions, n))
    
    # Gaussian noise on the y with a width equal to the 7.5% of the range 
    # of the function in the x interval
    if dimensions == 2:
        z = f(x)
    else:
        z = f(x, dimensions)
        
    range_f = abs(np.max(z) - np.min(z))
    width = 0.075 * range_f
    #print('Range: {:.3f} - Width: {:.3f}'.format(range_f, width))
    
    noise = width * np.random.choice([-1,1], size=n) * np.random.standard_normal(size=n)
    
    if dimensions == 2:
        z_noise = f(x) + noise
    else:
        z_noise = f(x, dimensions) + noise

    
    if (dimensions == 2) and visualize:
        # -------- plot noisy data ---------
        from mpl_toolkits.mplot3d import Axes3D

        plot3d = Axes3D(plt.figure()) # get handler for a 3d figure plot
        plot3d.view_init(elev=25., azim=65)
            
        # plot data points
        plot3d.scatter3D(x[0], x[1], z_noise, c=z_noise, cmap='Greens');
       
        #Plot the underlying three-dimensional plane
        xplane = np.linspace(xmin, xmax, 2000)
        yplane = np.linspace(xmin, xmax, 2000)
        
        xx, yy = np.meshgrid(range(xmin, xmax), range(xmin, xmax))
        xy_coord = np.array([xx, yy])
        z = f(xy_coord)
        plot3d.plot_surface(xx, yy, z, alpha=0.2)

        plot3d.set_xlabel('$x_1$')
        plot3d.set_ylabel('$x_2$')
        plot3d.set_zlabel('$f(x_1, x_2)$')

        plt.show()
    
    # store the data in a pandas DataFrame
    df = pd.DataFrame(data = x.T, columns = ['x{}'.format(i) for i in range(1, dimensions+1)])
    df['f(x)'] = z_noise
    return df

# this generate 'samples' examples based on a Gaussian noise applied 
# to the points of a given hyperplane
samples = 20
df = generate_multivariate_regression_data(hyperplane_2D, samples, visualize=True)
#df = generate_multivariate_regression_data(hyperplane_nD, samples, dimensions=40, visualize=False)
```

The ${\bf X}$ matrix and the ${\bf Y}$ vector are stored in the pandas dataframe `df`:

```python
df = df.round(2)
X = df.iloc[:, [0,1]]

# add the bias component
X.insert(0, 'x0', np.ones( (X.shape[0]) ))         
         
Y = df.iloc[:, -1]

print('X:\n',X)
print('\nY:\n',Y)
```

Let's compute the weights of the linear regression model by using the above formula:

$${\bf w}^* = ({\bf X}^T {\bf X})^{-1}{\bf X}^T{\bf Y}$$

### Rank of a matrix (Optional)

#### Linear independence (Optional)

A set of vectors in $\mathbb{R}^m$, ${\bf x}_1, {\bf x}_2, \ldots, {\bf x}_d$ is linearly independent if the equation of their *linear combination*

$$c_1{\bf x}_1 +c_2{\bf x}_2 + \ldots +c_d{\bf x}_d = 0$$

can be satisfied *only* for the (trivial) choice $c_1 = c_2 = \ldots \ c_d = 0$.

If this is the case, it means that there's not a choice of the coefficients $c_i$, apart from the trivial choice ${\bf c} = {\bf 0}$, that can allow to express one vector as a linear combination of the others. In other words, the vectors are **linearly 
independent.**

If it is possible to find a non-trivial choice of the coefficient, ${\bf c}$, the vectors are said **linearly dependent.**


Verifying linear independence amounts to find the solution of a system of linear equations, as it is shown in the example below from wikipedia:

<div>
  
</div>

In this case, $a_1, a_2, a_3$ are found as the solutions of the linear system:

$$\begin{cases}
\begin{matrix}
a_1 -3a_2 + 2a_3 = 0\\
a_1 + 2a_2 + 4a_3 = 0
\end{matrix}
\end{cases}
$$

The system has two equations and three unknowns. As it is expected, there are infinite solutions. E.g., the assignment $a_1=-16, a_2=-2, a_3=5$ satisfies the equations. 
In this case, the vectors are linearly dependent.

Note that, in general, <u>if there are more vectors, $d$, than dimensions, $m$, then the vectors are linearly dependent.</u> In the example, $d=3$ and $m=2$. The result is an **underdetermined system** (a system with fewer equations than unknowns), so that has usually infinite solutions.

There are various techniques for verifying the linear independence of vectors.

https://en.wikipedia.org/wiki/Linear_independence

#### Definition of rank (Optional)

<u>Rank of a matrix with $r$ rows and $c$ columns:</u> the number of row vectors or column column vector in the matrix that are linearly independent.

- If $r < c$ the **maximum rank** of the matrix is $r$ <br><p>

- If $r > c$ the **maximum rank** of the matrix is $c$

In other words, maximum (possible) rank is equal to the minimal dimension of the matrix, either in terms of rows or columns.
    
E.g., a $4\times 2$ matrix can have at most rank 2; a $4\times 5$ matrix can have at most rank 4.
    
A matrix has **full rank** if its rank is the same as the maximum rank.

<div>
  
</div>
Row 2 of matrix $A$ is equal to twice row 1. Therefore, rows 1 and 2 are linearly dependent. Matrix $A$ has only one linearly independent row, so its rank is 1. 
Hence, matrix $A$ is not full rank.

All rows of matrix $B$ are linearly independent, so the rank of matrix $B$ is 3. Matrix $B$ is full rank.

 $${\bf C} = {\begin{bmatrix}1&0&1\\-2&-3&1\\3&3&0\end{bmatrix}}$$
 
 The matrix $C$ has rank 2: the first two columns are linearly independent, so the rank is at least 2, but since the third is a linear combination of the first two (the second subtracted from the first), the three columns are linearly dependent so the rank must be less than 3.

#### Rank of the feature matrix (Optional)

Since (it can be proved that) for a generic matrix $A$ it holds that $rank(A) = rank(A^T A)$, in our regression problems it is the feature matrix ${\bf X}$ that determines the solution of the problem based on the expression

$${\bf w}^* = ({\bf X}^T {\bf X})^{-1}{\bf X}^T{\bf Y}$$

In our datasets, having two (or more) linearly dependent columns it means that the features are not unique, they have linear correlations that reduce the effective dimensionality of the problem representation.

### Rank of the feature matrix ${\bf X}$ vs. number of rows and of columns (Optional)

The feature (or design) matrix ${\bf X}$ has $n$ rows (examples) and $d$ columns (number of features), it is an $n\times d$ matrix.

The (Gram) matrix ${\bf X}^T{\bf X}$ is an $(d\times n)\cdot(n\times d) = d\times d$ matrix. 
    
The dimension of the vector ${\bf w}$ to be estimated is equal to $d$: it weights the impact of each feature determining the output value $y$.

<div style="text-align: left;">
  
</div>

The max rank of the matrix ${\bf X}^T{\bf X}$ is necessarily equal to $d$, which is the number of columns/features of the matrix ${\bf X}$.

Using this observation and based on the fact that $rank({\bf X}) = rank({\bf X}^T {\bf X})$, the following reasonings can be done about the rank of ${\bf X}$, and therefore of ${\bf X}^T {\bf X}.$

- If $n < d$ $\rightarrow$ $rank({\bf X}) \le n$ (rank of ${\bf X}$ is $n$, at most), and since $n$ is strictly less than $d$ $\rightarrow$ the matrix ${\bf X}^T {\bf X}$ cannot have full rank, meaning that the **matrix inversion operation can trigger an issue.** 

  This is the case when the feature matrix is *fat* (larger than taller) and the number of examples is less than the number of features (and therefore of the number of parameters ${\bf w}$ to determine). This is usually not the case in large datasets.<br><p>

- If $n > d$ $\rightarrow$ $rank({\bf X}) \le d$, such that the $d\times d$ matrix ${\bf X}^T {\bf X}$ *can* have full rank, given that  there are no linear dependencies among the column vectors of the matrix (which was indeed the case of the previous example).
